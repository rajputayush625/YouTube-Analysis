```python
import pandas as pd
```

## In this notebook, I'll show how to display the entire dataframe as a scrollable table
Keep in mind that displaying a large dataframe will cause your browser to freeze and crash. 

So if you're doing this on a large dataframe, try to only show a limited number of rows

```python
df = pd.read_csv("C:/Users/rajpu/Downloads/trending_yt_videos_113_countries.csv")
```

## Get a random sample of 500 rows.
So that i won't run out of RAM

```python
random_500_df = df.sample(500)
```

## Convert the random_500_df to html

```python
html_table = random_500_df.to_html(index=False)
```

## Prepare the html output as a div with custom height. 
You can change the height to show more or less rows at once

```python
html_output = """
<div style="height:600px; width:100%; overflow:auto;">
    {}
</div>
""".format(html_table)
```

```python
# html_output
```

## Show the html.
I'll be using IPython.display

> Use SHIFT + SCROLL for horizontal scrolling

```python
from IPython.display import display, HTML

display(HTML(html_output))
```
